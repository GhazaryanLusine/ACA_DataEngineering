```python
Guest_id = Guest['Guest_ID'].values
Na= ["Nan","Nan","Nan","Nan","Nan","Nan","Nan","Nan","Nan","Nan"]
Guest_id=np.concatenate((Guest_id, Na), axis=0)
```

```python
import datetime

def random_date_start():
    base = datetime.datetime.today()
    date=  base+datetime.timedelta(days=randint(1,15)) 
    return date.date()
def random_date_end():
    base = datetime.datetime.today()+datetime.timedelta(days=randint(1,14))
    date=  base+datetime.timedelta(days=randint(15,30)) 
    return date.date()
        
```

```python
def gen():
        rand=random.choice(Guest_id)
        if rand == "Nan":
            data=[randint(10000000,99999999),random.choice(Rooms['Room_id'].values),"Nan",random.choice(Tour_Agent['Tour_Agent'].values),random_date_start(),random_date_end()]
            return data
        else:
            data=[randint(10000000,99999999),random.choice(Rooms['Room_id'].values),rand, "Nan",random_date_start(),random_date_end()]
            return data
    
            
```

```python
def dataframe():
    df = pd.DataFrame(columns=["Booking_id","Room_id","Guest_id","Agent_id","Date_From","Date_to"])
    for i in range(2000):
        df.loc[i] = gen()
    return df
    
```

```python
Booking=dataframe()
```

```python
List=["Confirmed","Confirmed","Confirmed","Confirmed","Confirmed","Cancelled","Tentative","Tentative","Tentative","Tentative"]

def gen(i):
    return random.choice(List)
```

```python
def gen_booking_status():
    List1=Booking['Booking_id']
    frame = {'Booking_id': List1}
    result = pd.DataFrame(frame)
    result["Status"]=result["Booking_id"].apply(gen)
    return result

Booking_status=gen_booking_status()
Booking_status.tail()
```

```python
def gen_review():
    rand=randint(1,10)
    if rand <= 5:
        data=[random.choice(Booking_status_conf['Booking_id'].values),rand,random.choice(Room_review_not_happy['Description'].values)]
        return data
    else:
        data=[random.choice(Booking_status_conf['Booking_id'].values),rand,random.choice(Room_review_happy['Description'].values)]
        return data
```

```python
def dataframe_review():
    df = pd.DataFrame(columns=["Booking_id","Review_score","Review_text"])
    for i in range(200):
        df.loc[i] = gen_review()
    return df
```

```python
def gen_service():
    rand=random.choice(random.choice((Cars['Car_id'].values,Restaurants['Restaurant_id'].values,Tours['Tour_id'])))
    if rand.startswith('C'):
        data=[random.choice(Booking['Booking_id'].values),rand,"Car_renting"]
        return data
    elif rand.startswith('T'):
        data=[random.choice(Booking['Booking_id'].values),rand,'Tour_booking']
        return data
    else:
        data=[random.choice(Booking['Booking_id'].values),rand,'Restaurant_rservation']
        return data
```

```python
def dataframe_service():
    df = pd.DataFrame(columns=["Booking_id","Service_id","Service_type"])
    for i in range(1000):
        df.loc[i] = gen_service()
    return df
```

```python
Currency= ["USD","EUR"]
```

```python
from random import randrange
from datetime import timedelta

def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)

from datetime import datetime
d1 = datetime.strptime('1/1/2022' , '%m/%d/%Y').date()
d2 = datetime.strptime('2/1/2022', '%m/%d/%Y').date()
print(random_date(d1, d2))
```

```python
def gen_payment():
        data=[("P"+ str(randint(10000000,99999999))),random.choice(Booking_status_conf['Booking_id'].values),random_date(d1, d2),random.choice(Payment_type['Payment_type_id'].values),
             random.choice(Payment_status['Payment_status_id'].values),random.choice(Currency)]
        return data
 
```

```python
def dataframe_payment():
    df = pd.DataFrame(columns=["Payment_id","Booking_id","Date","Payment_type_id","Payment_status_id","Payment_Currency"])
    for i in range(1000):
        df.loc[i] = gen_payment()
        df.drop_duplicates(subset ="Booking_id",
                     keep = "first", inplace = True)
    return df
```
